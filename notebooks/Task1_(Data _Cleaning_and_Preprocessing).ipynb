## **Data Preprocessing**

#### Type Cunversion

```python
'''

Conversion of Data Time Column

'''

df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"], format="%d-%m-%Y")
```

```python
'''

Convert the Categorical Column

'''

categorical_cols = ["Education", "Marital_Status"]
for col in categorical_cols:
    df[col] = df[col].astype("category")
```

#### Handling Missing Values

```python
df["Income"] = pd.to_numeric(df["Income"], errors="coerce")
df["Income"].fillna(df["Income"].median(), inplace=True)
```

```python
print(df.info())
```

#### Column Name Cunversion

```python
df.columns = df.columns.str.strip().str.replace(" ", "_").str.lower()
```

## **Advance Preprocessing**

```python
df.head()
```

```python
X = df.drop(columns=["response"])
Y = df["response"]
```

```python
df.info()
```

```python
categorical_cols = X.select_dtypes(include=["category"]).columns.tolist()
numerical_cols = X.select_dtypes(include=["int64", "float64"]).columns.tolist()
```

```python
numeric_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])
```

```python
categorical_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(drop="first", handle_unknown="ignore"))
])
```

```python
preprocessor = ColumnTransformer(transformers=[
    ("num", numeric_pipeline, numerical_cols),
    ("cat", categorical_pipeline, categorical_cols)
])
```

```python
pipe = Pipeline(steps=[
    ("preprocessing", preprocessor)
])
```

```python
processed_data = pipe.fit_transform(X)
```

```python
processed_data.shape
```

```python
numeric_features = numerical_cols
encoded_categorical_features = pipe.named_steps["preprocessing"].named_transformers_["cat"].named_steps["encoder"].get_feature_names_out(categorical_cols)
```

```python
all_feature_names = list(numeric_features) + list(encoded_categorical_features)
```

```python
X_df = pd.DataFrame(processed_data, columns=all_feature_names)
X_df.head()
```
